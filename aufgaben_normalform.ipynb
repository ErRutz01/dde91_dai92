```python
FILE_PATH = os.path.join("raw_data")
```

```python
DATA = os.path.join(FILE_PATH, "Book_Buddy.csv")
```

```python
df = pd.read_csv(DATA, sep=";")
```

```python
df_customer = df[["CustomerID"]].copy()

for a in df["CustomerAddress"]:
    spliter = a.split(",")
    df_customer["street"] = spliter[0]
    df_customer["city"] = spliter[1]
    df_customer["zip"] = spliter[2]

df_customer.head(3)
```
